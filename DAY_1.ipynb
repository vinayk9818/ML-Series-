# Based on Learinig Type (Important)
1️⃣ **Batch Learning**



1 The model is trained using the entire dataset at once.

2 After training, the model is fixed.

3 If new data arrives, the model must be retrained from scratch.

**When it is used:**

When data is stable

When frequent updates are not required

**Limitation:**

Expensive for very large datasets

Cannot adapt quickly to new data

2️⃣ **Online Learning**

1 The model is trained incrementally, using small batches of data.

2 It can learn continuously as new data arrives.

3 It updates itself step by step.

**When it is used:**

When data arrives continuously

For real-time systems

**Example:**

Stock market prediction

Recommendation systems

Real-time fraud detection




# Based on Generaliztion
ML systems are also classified by how they make predictions on new data.

1️⃣ Instance-Based Learning

👉 The system learns by remembering examples.

👉 When a new instance comes, it compares it with stored examples.

👉 It predicts based on similarity.

**How it works:**

Memorize training data

Use a similarity measure (like distance)

Find nearest examples

Make prediction

**Example:**

k-Nearest Neighbors (k-NN)

Spam detection using similarity


# Main Challenges of Machine Learning

**1️⃣ `Insufficient Quantity of Training Data`**

Not enough data → model cannot learn properly.

More high-quality data usually improves performance.

`reseacher paper of Microsoft` : unreasonable effectness of data

 2️⃣**Non-Representative Training Data**

Training data must represent real-world cases.

If not → model will not generalize well.

**3️⃣ Poor-Quality Data**

Noisy data

Missing values

Outliers
→ Leads to poor predictions.

**4️⃣ Irrelevant Features**

Not all input features are useful.

Feature selection and feature engineering are important.

**5️⃣ Overfitting (Most Important )**

Model performs well on training data

But performs poorly on new data
