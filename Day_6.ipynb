# **Important Terms**
1. TP (True Positive) : Actual positive tha → model ne positive bola

2. True Negative (TN) :
The model predicted negative, and it was actually negative.

3. False Positive (FP) :
The model predicted positive, but it was actually negative.

4. False Negative (FN) :
The model predicted negative, but it was actually positive.

# **Metrics Derived from Confusion Matrix**

**Accuracy**

* (TP+TN)/Total

**Precision**

* TP/(TP+FP)

Measures: Out of all predicted positives, how many were correct?

**Recall**

* TP/(TP+FN)

Measures: Out of all actual positives, how many did we correctly detect?

# **3. Percision and Recall Metrics**

**Precision and Recall are not just model metrics — they depend on the goal of the problem and the cost of mistakes**

**🟢 When to Focus on Precision**

Use Precision when False Positives are costly
(When predicting “positive” incorrectly causes serious problems.)

Common Problem Types:

1. Spam Detection (marking an important email as spam ❌)

2. Fraud Detection (flagging an innocent user as fraud ❌)


Core Thinking:

* If I say something is positive, I must be very sure.”


**🔵 When to Focus on Recall**

Use Recall when False Negatives are costly
(Missing a real positive case is dangerous.)



Common Problem Types:

1. Cancer Detection (missing a patient who actually has cancer ❌)

2. Disease Prediction

Core Thinking:

* I cannot afford to miss any real positive case


**If False Positive cost > False Negative cost → Focus on Precision**

**If False Negative cost > False Positive cost → Focus on Recall**


# **Percision And Recall Tradeoff**

Precision and Recall move in opposite directions when we adjust the decision threshold of the same model.

If you increase one, the other usually decreases.

image.png







# **Important points regrading Tradeoff**

This tradeoff discussion applies:

✅ When the model is the same

✅ When we are adjusting the decision threshold

If we:

* Change features

* Change algorithm

* Change hyperparameters

* Add more data

**Important Observation**

* Lower threshold → High Recall, Low Precision

* Higher threshold → High Precision, Low Recall

* Tradeoff mainly occurs when the model stays the same and only threshold changes

* Changing features/model can shift the whole curve upward


# **ROC - AUC Curve**

It is a curve that shows how well a classification model separates classes at different thresholds.

ROC does NOT depend on one fixed threshold.
It evaluates performance across all thresholds.

**By analyzing the ROC curve, we can choose the best threshold that balances sensitivity (Recall) and false positive rate according to the problem’s objective**

image.png


Since a better classifier aims to achieve high true positive rate and low false positive rate, we select the threshold that lies closest to the top-left corner of the ROC curve, as it satisfies this optimal trade-off condition


ROC does NOT directly optimize precision.
It optimizes the balance between:

* Sensitivity (Recall)

* False Positive Rate

If you want to optimize Precision–Recall specifically,
then Precision–Recall curve is more appropriate.


# **Error Analysis**

Systematically analyzing where and why the model is making mistakes in order to improve it.

Step 1: Compare True vs Predicted Labels

Step 2: Plot the Confusion Matrix

Step 4: Identify the Root Cause

Final Understanding

Accuracy tells us how many mistakes the model made.
Error analysis tells us why those mistakes happened.

Improving a machine learning model requires understanding the root cause of errors, not just increasing performance metrics.
