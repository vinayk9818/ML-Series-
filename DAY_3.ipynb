# pros And Cons of Linear Regression


| Category | Point                              | Explanation (SDE Level Insight)                                                   |
| -------- | ---------------------------------- | --------------------------------------------------------------------------------- |
| ✅ Pros   | Simple & Easy                      | Easy to implement, understand, and debug. Good for whiteboard interviews.         |
| ✅ Pros   | Interpretable                      | Coefficients clearly show feature impact on output. Useful in business analytics. |
| ✅ Pros   | Computationally Efficient          | Fast training on small/medium datasets. Low computational cost.                   |
| ✅ Pros   | Convex Optimization                | MSE loss is convex → global minimum guaranteed.                                   |
| ✅ Pros   | Good Baseline Model                | Often used as first model before trying complex ML algorithms.                    |
| ❌ Cons   | Only Captures Linear Relationships | Cannot model nonlinear patterns unless transformed.                               |
| ❌ Cons   | Sensitive to Outliers              | Squared error amplifies extreme values.                                           |
| ❌ Cons   | Assumption Heavy                   | Requires linearity, no multicollinearity, homoscedasticity, etc.                  |
| ❌ Cons   | Multicollinearity Issue            | Highly correlated features make weights unstable.                                 |
| ❌ Cons   | Weak for Complex Data              | Not suitable for images, text, audio, or high-dimensional nonlinear data.         |


# **Impletation Linear Regression using sklearn**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv("/content/test.csv")
df.sample(5)
```

```python
df.columns
```

```python
df.info()
df.describe()
```

```python
plt.scatter(df['x'],df['y'])
plt.xlabel("CPGA")
plt.ylabel("Placement")
plt.title("CGPA VS placement")
```

```python
x=df.iloc[:,:1]
y=df.iloc[:,-1]
```

```python
x.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
# check accuray using r2 and mse score
r2=r2_score(X_test,model.predict(X_test))
print(f"model r2 accuracy is : {r2}")
```

```python
mse=mean_squared_error(X_test,model.predict(X_test))
print(f"model error is : {mse}")
```

```python
train_error=model.score(X_train,y_train)
test_error=model.score(X_test,y_test)
print(f"training error is {train_error}")
print(f"test error is {test_error}")
```

**from the above bolck of code its clear that if model is do goog in both trainig and as welllas test then model is good**

```python
plt.scatter(df['x'],df['y'])
plt.xlabel("x-asix")
plt.ylabel("y-asix")
plt.title("Linear Regression ")
plt.plot(X_test,model.predict(X_test),color='Red')
```
